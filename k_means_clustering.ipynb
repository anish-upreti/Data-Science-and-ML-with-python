## KMeans Clustering

K-Means Clustering: A Brief Overview
K-Means is one of the most popular unsupervised machine learning algorithms used for clustering. It groups similar data points into k clusters based on feature similarity.

The algorithm works by iteratively minimizing the distance between data points and the centroids (mean points) of the clusters.

Steps in K-Means Clustering:  
Choose the number of clusters (k): You specify k, the number of clusters you want to find. This is often chosen based on prior knowledge or through methods like the elbow method.

Initialize Centroids: Randomly choose k data points as the initial centroids of the clusters. These centroids act as the "center" of each cluster.

Assign each point to the nearest centroid: Each data point is assigned to the cluster whose centroid is nearest (typically using Euclidean distance).

Recalculate the centroids: After assigning all data points to clusters, the algorithm recalculates the centroids by taking the mean of all points in each cluster.

Repeat: Steps 3 and 4 are repeated until convergence â€” when centroids no longer change significantly or the algorithm reaches a specified number of iterations.

Output: The algorithm outputs the cluster labels for each data point and the final centroids of the k clusters.

```python
## For 2D data points

## importing libraries and modules
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

## manually create a numpy array with 2D data points (x, y)
X = np.array([
    [1.0, 2.0],
    [1.5, 1.8],
    [5.0, 8.0],
    [8.0, 8.0],
    [1.0, 0.6],
    [9.0, 11.0],
    [8.0, 2.0],
    [10.0, 2.0],
    [9.0, 3.0],
    [2.0, 3.0]
    ])

## apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=11)
kmeans.fit(X)

# Predict which cluster each point belongs to
y_kmeans = kmeans.predict(X)  # This assigns each point to a cluster label (0, 1, or 2).
                            # y_kmeans = kmeans.predict(X) and kmeans.labels_ give you the same result after fitting, but for different purposes.
# get the cluster centers
centers = kmeans.cluster_centers_

print("y_kmeans: ", y_kmeans)
print("Centers: \n", centers)

## visualize the results
plt.scatter(X[:,0], X[:, 1], c=y_kmeans, cmap = "viridis", s = 100, label="cluster points")  # X[:, 0] = x-values, X[:, 1] = y-values.
                                                                    # c=y_kmeans: colors the points based on their cluster assignment.
plt.scatter(centers[:, 0], centers[:, 1], c="red", s = 200, alpha=0.75, marker="X", label="Centroids")

plt.title("K-Means Clustering on Custom NumPy Data")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.show()


```

```python
## For 1D  data

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 1D data (wrapped as 2D array for sklearn)
X = np.array([[150], [155], [160], [162], [165], [170], [172], [180], [185], [190]])

# Apply KMeans with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)

# Predicted cluster labels and centers
labels = kmeans.labels_          # contains the cluster labels assigned to each data point after the model has been fitted.
centers = kmeans.cluster_centers_

# Plotting 1D points along x-axis, setting y=0
plt.scatter(X[:, 0], np.zeros_like(X), c=labels, cmap='viridis', s=100, label="cluster points")
plt.scatter(centers[:, 0], np.zeros_like(centers), c='red', s=200, marker='X', label='Centers')

plt.title("K-Means Clustering on 1D Data")
plt.xlabel("Value")
# plt.yticks([])  # Hide y-axis
plt.legend()
plt.grid(True)
plt.show()
```

```python
## For 3D data
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D   ## Required for 3D plotting

# Create synthetic 3D data (x, y, z)
X = np.array([
    [1, 2, 1],
    [1.5, 1.8, 1.2],
    [5, 8, 6],
    [8, 8, 7],
    [1, 0.6, 0.5],
    [9, 11, 10],
    [8, 2, 2],
    [10, 2, 3],
    [9, 3, 3.5],
    [2, 3, 1.5]
])

# apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# plotting in 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")

# plot the each cluster with differennt color
colors = ["red", "blue", "green"]
for cluster_id in range(3):
    cluster_points = X[labels == cluster_id]
    ax.scatter(cluster_points[:,0], cluster_points[:,1], cluster_points[:,2],
               c=colors[cluster_id], label = f"cluster {cluster_id}", s= 50)
    
#plot the cluster centers
ax.scatter(centers[:,0], centers[:,1], centers[:,2], c="purple", marker="X", label="centers")

#plot formatting
ax.set_title("KMeans Clustering in 3D")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.legend()
plt.tight_layout()
plt.show()
```
