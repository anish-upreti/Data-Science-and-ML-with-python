### SpeechRecognition
The SpeechRecognition library in Python is a popular tool used to convert spoken language into text using speech recognition engines. It's user-friendly and supports multiple recognition engines and APIs, such as Google Web Speech API, CMU Sphinx (offline), Microsoft Bing Voice Recognition, and others.  

Key Features:  
Easy-to-use API for recognizing speech from audio files or live microphone input.

Supports multiple engines, including both online and offline options.

Works with various audio formats like WAV, AIFF, and FLAC.

Real-time speech recognition with microphone input.

```python
import speech_recognition as sr

# Initializing recognizer class
r = sr.Recognizer()

# Reading audio file as source
with sr.AudioFile("input.wav") as source:
    # audio_text = r.listen(source) ## mostly used when capturing live speech from a microphone.
    audio_text = r.record(source)   # mostly used when processing pre-recorded audio from a file.
    try:
        # using google speech recognition
        text = r.recognize_google(audio_text)
        print("Converting audio into text\n")
        print(text)
    except:
        print("sorry run again")
    
# Adding English Language option
text = r.recognize_google(audio_text, language="en-GB")
print(text)
```
