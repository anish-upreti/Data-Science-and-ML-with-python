```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('file.csv')

# Plot the runs comparison between the two teams across overs (line plot or line graph)
plt.plot(df['Overs'], df['SA_Runs'], label='South Africa', marker='o', color='blue', linestyle='-', linewidth=2)
plt.plot(df['Overs'], df['NP_Runs'], label='Nepal', marker='o', color='red', linestyle='-', linewidth=2)

# Add labels, title, and legend
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.title('Runs Comparison Between Nepal and South Africa Across Overs')
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
