## What is Linear Regression?
Linear Regression is a statistical method used to model the relationship between a dependent variable (target) and one or more independent variables (predictors) by fitting a linear equation to observed data.

Types:  
Simple Linear Regression – One independent variable

Multiple Linear Regression – Two or more independent variables  


## Key Assumptions:
Linearity of the relationship

Independence of errors

Homoscedasticity (constant variance of errors)  

        Homoscedasticity:  
        This refers to the constant variance of errors (residuals) across all levels of the independent variable(s).

        Homoscedasticity:
        Error spread is the same everywhere.

        Example: Points are randomly scattered around the regression line with equal spread.

Normality of error terms  

        The error terms (residuals) — the differences between the actual and predicted values — should follow a normal distribution:

        This does not mean the data or the dependent variable must be normal — only the errors.

No multicollinearity (for multiple regression)  

        Multicollinearity occurs when two or more independent variables are highly correlated with each other.

        This means they provide redundant information, making it hard to determine the individual effect of each variable on the dependent variable.

## How It Works:
Fits a line that minimizes the sum of squared errors (residuals) between predicted and actual values using Ordinary Least Squares (OLS).  

OLS (Ordinary Least Squares):
OLS is the most common method for estimating the coefficients in linear regression.

Goal:
To minimize the sum of the squared differences (errors) between the actual values and the predicted values.
 
So it finds the best-fitting line that makes the total squared prediction error as small as possible.

## Performance Metrics:
R² (R-squared): Proportion of variance explained


MSE/ RMSE: Measures average error magnitude

## Applications:
Predicting values (e.g., sales, prices, scores)

Understanding variable relationships

Trend analysis

## Pros:
Simple and interpretable

Fast to train

## Cons:
Assumes linearity

Sensitive to outliers

Poor performance with non-linear data



## ✅ What is Logistic Regression?
Logistic Regression is a statistical method used for binary classification — i.e., predicting whether something belongs to Class 0 or Class 1 (e.g., spam or not spam, sick or healthy).

Unlike linear regression, it predicts probabilities using the logistic (sigmoid) function.

  

## Use Cases:
Spam detection

Disease prediction

Customer churn

Loan default prediction

## Types:
Binary Logistic Regression – Two classes (0 or 1)

Multinomial Logistic Regression – More than two unordered categories

Ordinal Logistic Regression – More than two ordered categories

## Assumptions:
Dependent variable is binary

Observations are independent

Little to no multicollinearity

Linearity between independent variables and log-odds of the outcome

Large sample size preferred

## Evaluation Metrics:
Accuracy

Precision, Recall, F1-score

Confusion Matrix

ROC Curve and AUC

## Pros:
Fast and simple

Outputs probabilities

Good for linearly separable data

##  Cons:
Assumes linearity in log-odds

Not suitable for complex relationships

Sensitive to outliers  

## Important formulas


## Linear regression vs Logistic Regression

| **Aspect**               | **Linear Regression**                                    | **Logistic Regression**                                    |
| ------------------------ | -------------------------------------------------------- | ---------------------------------------------------------- |
| **Purpose**              | Predict **continuous** values                            | Predict **probabilities / classes** (usually binary)       |
| **Output Range**         | $(-\infty, +\infty)$                                     | $(0, 1)$ (probabilities)                                   |
| **Target Variable**      | Continuous (e.g., price, weight)                         | Categorical (e.g., yes/no, 0/1)                            |
| **Equation Form**        | $y = \beta_0 + \beta_1x + \cdots + \beta_nx_n$           | $p = \frac{1}{1 + e^{-(\beta_0 + \beta_1x + \cdots)}}$     |
| **Loss Function**        | Mean Squared Error (MSE)                                 | Log Loss (Binary Cross-Entropy)                            |
| **Function Used**        | Linear function                                          | Sigmoid (logistic) function                                |
| **Relationship Modeled** | Direct linear relationship between inputs and output     | Linear relationship between inputs and **log-odds**        |
| **Use Cases**            | Predicting quantities (e.g., house price, salary)        | Classification (e.g., spam detection, disease prediction)  |
| **Interpretability**     | Coefficients show direct change in output per unit input | Coefficients show change in **log-odds** or **odds ratio** |
| **Error Assumptions**    | Normally distributed residuals                           | No assumption of normality for dependent variable          |


```python
## Simple Linear Regression Implementation
from sklearn.linear_model import LinearRegression
import numpy as np

# data
x = [1, 2, 3, 4, 5]  #study time in hours
y = [50, 65, 70, 80, 85]  #marks

# converting x to 2D array
X = np.array(x).reshape(-1, 1)   # for regression tasks, input are expected in this shape: (n_samples, n_features)
Y = np.array(y)                  # ouput or target is expected to be of shape: (n_samples,)
 
# create and train the model
model = LinearRegression()
model.fit(X, Y)

# get the slope(coefficient) and the intercept
m = model.coef_[0]     #slope
c = model.intercept_    #intercept
print(f"Slope: {m}")
print(f"Intercept: {c}")

# final equation
print(f"Final equation of line is y = {m:.2f}x + {c:.2f}")

# predicting y when x = 6
x_test = np.array([[6]])
y_pred = model.predict(x_test)
print(f"Prediction: If x = 6, then y = {y_pred[0]:.0f} marks")   #y_pred is a numpy array
```

```python
# for visualization
import matplotlib.pyplot as plt

# plotting the original data
plt.scatter(X, Y, color = "Blue", label = "Actual data (study hours vs marks)")

# plotting the regression line
y_line = model.predict(X)
plt.plot(X, y_line, color = "red", label = "Regression Line")

# Labels and Title
plt.title("Simple Linear Regression")
plt.xlabel("Study Time (hours)")
plt.ylabel("Marks")
plt.legend()

plt.show()
```

```python
## Simple example of Multiple Linear Regression
from sklearn.linear_model import LinearRegression
import numpy as np

# Example data (study time, sleep hours) → predict marks
# Each row: [study hours, sleep hours]
X = np.array([
    [1, 6],
    [2, 7],
    [3, 6.5],
    [4, 8],
    [5, 7]
])
y = np.array([50, 60, 65, 75, 85])  # marks

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Model parameters
coefficients = model.coef_   # array of slopes for each feature
intercept = model.intercept_
print(f"Intercept: {intercept:.2f}")
print(f"Coefficients: {coefficients}")

# Prediction
x_test = np.array([[6, 7]])  # 6 hrs study, 7 hrs sleep
y_pred = model.predict(x_test)
print(f"Prediction for [6 hrs study, 7 hrs sleep]: {y_pred[0]:.2f} marks")
```

```python
## Implementing logistic regresssion
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
data = load_iris()
X = data.data  # Features (sepal length, sepal width, petal length, petal width)
y = data.target  # Target variable (species)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train the Logistic Regression model
model = LogisticRegression(max_iter=200)  # max_iter to ensure convergence
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Plotting the Confusion Matrix using Seaborn heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=data.target_names, yticklabels=data.target_names)
plt.title('Confusion Matrix for Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Predicting probabilities for each class
y_prob = model.predict_proba(X_test)
print("Predicted probabilities for each class:")
print(y_prob)

# Making a single prediction
single_prediction = model.predict([X_test[0]])
print(f"Prediction for single sample: {data.target_names[single_prediction][0]}")
```
