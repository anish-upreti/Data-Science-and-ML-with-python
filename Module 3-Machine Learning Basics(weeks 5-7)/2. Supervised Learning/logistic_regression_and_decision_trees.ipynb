```python
## logistic regression
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression

x, y = load_iris(return_X_y=True)

print(x.shape)  # (150, 4)
print(y.shape)  # (150,)
print(set(y))    # [0 0 0 0 0]
```

## ✅ model.score(x, y)

This evaluates the model on the data it was trained on.

It shows how well the model fits the training data.

⚠️ Warning: This score can be misleadingly high, because the model has already "seen" this data. It might be overfitting.

```python
## splitting and training the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state=12)

#initialize the model
model = LogisticRegression(max_iter=100) ## may take more iterations to converge
model.fit(x_train, y_train)

#predict 
y_pred = model.predict(x_test)

#evaluate
accuracy = model.score(x_test, y_test)
print(f"Accuracy: {accuracy:.2f}")
```

## ✅ accuracy_score(y_test, y_pred):
This function:

Compares the true labels (y_test) with the model’s predicted labels (y_pred)

Calculates the percentage of correct predictions  

Formula is:  
Accuracy= 
Total number of predictions/Number of correct predictions

​


```python
## can also use the following for accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)   # 	Explicit and flexible (best for metrics)
print(f"Accuracy: {accuracy:.2f}")
```

```python
## Decision Trees
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# initialize the model
dts = DecisionTreeClassifier()
dts.fit(x, y)

# visualize the tree
tree.plot_tree(dts)
plt.title("Decision Trees")
```
