```python
# importing libraries and csv file
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("weather.csv")
df
```

```python
months = df["Month"]
high = df["High"]
low = df["Low"]
rainfall = df["Rainfall"]

table = [rainfall, high, low]
df = pd.DataFrame(table, index = ["Rainfall", "High", "Low"])
df
```

```python
## plotting
plt.figure(figsize=(20,10))                # Set the figure size to 20x10 inches
plt.plot(high, "s-b", linewidth=3)        # Plot 'high' with square markers in blue and thick lines
plt.plot(low, "s-r", linewidth=3)         # Plot 'low' with square markers in red and thick lines
plt.plot(rainfall, "-g", linewidth=3)     # Plot 'rainfall' as a green line
plt.fill_between(months, rainfall, color="green", alpha=0.2)  # Fill area under 'rainfall' curve

#Labels and grid
plt.ylabel("Temperature")
plt.grid(axis="y")
plt.title("Weather of Kathmandu (in celcius)")
#Add legend
plt.legend(["Max Temperature", "Min Temperature", "Rainfall in cm"])

# use xticks
plt.xticks([])

# display table beneath the plot
plt.table(cellText=table, rowLabels=["Rainfall in cm", "Max Temperature", "Min Temperature"])


plt.show()                                # Display the plot
```
